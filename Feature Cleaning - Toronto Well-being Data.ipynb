```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv("wellbeing_toronto.csv")
df.head()
```

```python
df['Pop 25-45 years'] = df['Pop 25 - 29 years'] + df['Pop 30 - 34 years'] + df['Pop 35 - 39 years'] + df['Pop 40 - 44 years']
df['Crime'] = df['Arsons'] + df['Murders'] + df['Thefts'] + df['Vehicle Thefts'] + df['Robberies'] + df['Break & Enters'] + df['Sexual Assaults'] + df['Assaults'] + df['Drug Arrests']

del df['Pop 25 - 29 years'] 
del df['Pop 30 - 34 years'] 
del df['Pop 35 - 39 years'] 
del df['Pop 40 - 44 years']

df = df.drop(df.columns[[6,7,8,9,10,11,12,13,14]], axis=1)
del df['Combined Indicators']
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="my_application",
                            format_string="%s, Toronto Ontario")

def add_lat(neigh):
    try:
        location = geolocator.geocode(neigh)
        lat = location.latitude
    except:
        lat = 99999
        
    return lat



def add_lng(neigh):
    try:
        location = geolocator.geocode(neigh)
        lng = location.longitude
    except:
        lng = 99999
        
    return lng
```

```python
df['Lat'] = df.apply(lambda x: add_lat(x['Neighbourhood']),axis=1)
```

```python
df[df['Lat'] == 99999].shape
```

```python
df['Lng'] = df.apply(lambda x: add_lng(x['Neighbourhood']),axis=1)
```

```python
df[df['Lng'] == 99999].shape
```

```python
df_again = df[(df['Lng'] == 99999) | (df['Lat'] == 99999)]
df_again.shape
```

```python
df_again['Lat'] = df_again.apply(lambda x: add_lat(x['Neighbourhood']),axis=1)
```

```python
df_again['Lng'] = df_again.apply(lambda x: add_lng(x['Neighbourhood']),axis=1)
```

```python
print(df_again[df_again['Lat'] == 99999].shape)
print(df_again[df_again['Lng'] == 99999].shape)
```

```python
df_new = df_again[(df_again['Lat'] != 99999) | (df_again['Lng'] != 99999)]
```

```python
df.drop(df[(df['Lng'] == 99999) | (df['Lat'] == 99999)].index, inplace=True)
```

```python
df_final = pd.concat([df, df_new])
df_final.reset_index(drop=True)
```

```python
df_final.reset_index(drop=True).to_csv('wellbeing_toronto_CLEANED.csv') 
```
